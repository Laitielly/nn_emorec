```python
import pickle
import numpy as np

from math import ceil
```

```python
# video_mae_train = 'new_embs_mae/train_mae_encdec.pickle'
# video_mae_val = 'new_embs_mae/val_mae_encdec.pickle'
video_mae_train = 'new_embs_mae/train_mae_enc.pickle'
video_mae_val = 'new_embs_mae/val_mae_enc.pickle'
```

```python
fea = pickle.load(open('features_newvf2.pickle', 'rb'))
video_maet_fea = pickle.load(open(video_mae_train, 'rb'))
video_maev_fea = pickle.load(open(video_mae_val, 'rb'))
```

```python
len(video_maev_fea)+len(video_maet_fea)
```

```python
len(set(list(video_maev_fea.keys())+list(video_maet_fea.keys())))
```

```python
video_mae = {**video_maet_fea, **video_maev_fea}
len(video_mae)
```

```python
len(fea), len(video_mae)
```

```python
from tqdm import tqdm

nn = {}
for i in tqdm(video_mae.keys()):
    if i not in nn:
        nn[i] = {}
    nn[i] = fea[i]
    nn[i]['mae'] = video_mae[i]
```

```python
len(nn)
```

```python
for i in nn.keys():
    print(nn[i], i)
    break
```

```python
with open('features_with_mae_enc.pickle', 'wb') as handle:
    pickle.dump(nn, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
